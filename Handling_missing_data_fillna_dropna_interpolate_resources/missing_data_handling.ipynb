Handling missing data -  fillna, interpolate, dropna

```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
type(df.temperature[0])
```

```python
df.set_index('day',inplace=True)
df
```

fillna-
fill all NaN with one spicific value

```python
new_df = df.fillna(0)
new_df
```

Fill na using column names and dict

```python
new_df = df.fillna({
        'temperature': df.temperature.mean(),
        'windspeed': df.windspeed.mean(),
        'event': 'No Event'
    })
new_df
```

Use method to determine how to fill na values

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

Use of axis - axis is either "index" or "columns"

```python
new_df = df.fillna(method="bfill", axis="columns") 
new_df
```

Limit Parameter

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

INTERPOLATE

```python
new_df = df.interpolate()
new_df
```

dropna

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
